## Complete Feature Engineering Pipeline

### Step 1: Import Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

### Step 2: Load Dataset

```python
df = pd.read_csv("employee_data.csv")
df.head()
```

### Step 3: Initial Data Analysis

```python
df.info()
df.describe()
```

### Step 4 Check Missing Values

```python
df.isnull().sum()
```

### Step 5 Handling Missing Values (Imputer)

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='mean')

df['Age'] = imputer.fit_transform(df[['Age']])
```

Explanation

- SimpleImputer fills missing values

- Strategy = mean (you can use median also)

```python
df.isnull().sum()
```

### Step 6 Encoding

- Machine Learning models cannot understand text (categorical data)
- They only understand numbers
- Encoding Converts Categorical Values Into Numerics Values


### Label Encoding

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['Gender'] = le.fit_transform(df['Gender'])
```

Explanation

- Male → 1, Female → 0

- Used for binary categories

### One Hot Encoding

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(drop='first', sparse_output=False)

encoded = ohe.fit_transform(df[['Department','City']])

encoded_df = pd.DataFrame(encoded, columns=ohe.get_feature_names_out())

df = pd.concat([df, encoded_df], axis=1)

df.drop(['Department','City'], axis=1, inplace=True)
```

### Step 7 Time-Based Feature Engineering

```python
df['Joining_Date'] = pd.to_datetime(df['Joining_Date'], format='%d-%m-%Y')

df['Year'] = df['Joining_Date'].dt.year
df['Month'] = df['Joining_Date'].dt.month

df['Years_at_Company'] = 2025 - df['Year']
```

Explanation

We extracted:

- Year

- Month

- Experience in company

- Dates → useful numeric features

### Step 8 Feature Scaling

- Features have different ranges
- Model gives more importance to bigger values
- So we scale to same range

#### Min-Max Scaling

```python
from sklearn.preprocessing import MinMaxScaler

minmax = MinMaxScaler()

df[['Age']] = minmax.fit_transform(df[['Age']])
```

- Range: 0 to 1

#### Standardization

```python
from sklearn.preprocessing import StandardScaler

std = StandardScaler()

df[['Experience']] = std.fit_transform(df[['Experience']])
```

- Mean = 0, Std = 1

#### Robust Scaling

```python
from sklearn.preprocessing import RobustScaler

robust = RobustScaler()

df[['Years_at_Company']] = robust.fit_transform(df[['Years_at_Company']])
```

- Best when outliers present

### Step 9 Skewed Data Handling

```python
print(df['Salary'].skew())
```

### Apply Log Transformation

```python
import numpy as np

df['Salary_log'] = np.log1p(df['Salary'])
```

Explanation

- Skewed data → uneven distribution

- Log transform → makes it normal

- Helps ML model perform better
